```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver

options = Options()
options.headless = True # "headless" for Selenium means no graphical user interface (GUI)
service = Service(executable_path="chromium.chromedriver")
b = webdriver.Chrome(options=options, service=service)
```

```python
from IPython.core.display import Image, display

def show_screen():
    b.set_window_size(200, 300)
    b.save_screenshot("screen.png")
    display(Image("screen.png"))
```

```python
import time

url = "https://tyler.caraza-harter.com/cs320/tricky/page1.html"
b.get(url)

attempts = 20
tbl = None

# this strategy is called "polling"
for _ in range(attempts):
    show_screen()
    try:
        tbl = b.find_element(by="id", value="coords")
        break
    except NoSuchElementException:
        print("NOT FOUND YET")
        time.sleep(0.2)

assert tbl != None
tbl
```
