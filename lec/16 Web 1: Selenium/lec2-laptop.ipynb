```python
import time

url = "https://tyler.caraza-harter.com/cs320/tricky/page1.html"
b.get(url)

attempts = 20
tbl = None

# this strategy is called "polling"
for _ in range(attempts):
    try:
        tbl = b.find_element(by="id", value="coords")
        break
    except NoSuchElementException:
        print("NOT FOUND YET")
        time.sleep(0.2)

assert tbl != None
tbl
```

```python
def grab_table(tbl):
    trs = tbl.find_elements(by="tag name", value="tr")
    rows = []
    for tr in trs:
        cells = tr.find_elements(by="tag name", value="td")
        cells = [cell.text for cell in cells]
        rows.append(cells)
    return pd.DataFrame(rows[1:], columns=rows[0])

grab_table(tbl)
```

# Click Example

```python
url = "https://tyler.caraza-harter.com/cs320/tricky/page2.html"
b.get(url)

while True:
    time.sleep(0.5)
    try:
        btn = b.find_element(by="id", value="more")
        btn.click()
    except NoSuchElementException:
        break

grab_table(b.find_element(by="tag name", value="table"))
```
