## Paths

A **path** is a sequence of edges to get from one node to another

````python
g = Graph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "E")
g.edge("E", "F")
g.edge("F", "A")
g.edge("X", "Y")
g
# is there a path between two nodes?  (from A to C, yes; from A to X, no)
# what is the shortest path between two nodes (A=>B=>C is shorter than A=>F=>E=>D=>C)
# "cycle" is basically a path that loops (e.g., A=>F=>E=>D=>C=>B=>A, but not X=>Y=>X)
# "connected" means there's a path between any two nodes
````

## Directed Graphs, DAGs

````python
from graphviz import Digraph # "directed graph"
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "A")
g
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "A")
g
# A is it's own great grandparent
# not a DAG, because there is at least one cycle
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("A", "C")
g.edge("B", "D")
g.edge("C", "D")
g
# weakly connected: path between any two nodes, ignore edge direction
# strongly connected (connected): path between any two nodes, respecting edge direction

# the following weakly connected, but it is not connected

# this is a dag, because it is directed and has no cycles (it's "acyclic")
````

````python
# "parent" (source of an edge): A is a parent of B and C
# "child" (destination of an edge): C is a child of A
# A is a "grandparent" of D
# A, B, and C are anscestors of D
````

## Trees

**Tree**: DAG where one node (the **root**) has no parents and all others have exactly one parent

````python
g = Digraph()
g.edge("A", "B")
g.edge("A", "C")
g.edge("C", "D")
g.edge("D", "E")
g.edge("D", "F")
g
# root: node with no parents (A)
# leaf: node with no children (B, E, F)
````

````python
# linked list: a tree, where every node has AT MOST one child
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g
````

# Binary Tree Example

````python
# class Edge:
#     pass

class Node:
    def __init__(self, name):
        self.name = name
        
        # linked list:
        # self.child = None
        
        # tree in general
        #self.children = [...nodes here...]
        
        # binary tree
        self.left = None
        self.right = None
        
    def dump(self, indent):
        print("   "*indent + self.name)        
        if self.left != None:
            self.left.dump(indent+1)
        if self.right != None:
            self.right.dump(indent+1)
        
A = Node("A")
B = Node("B")
C = Node("C")
D = Node("D")
A.left = B
A.right = C
A.right.right = D

A.dump(indent=0)
````
