## Paths

A **path** is a sequence of edges to get from one node to another

````python
g = Graph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "E")
g.edge("E", "F")
g.edge("F", "A")
g.edge("X", "Y")
g
# is there a path between two nodes?  (yes from A to C; no from A to X)
# what is the shortest path between two nodes?  (A=>B=>C is a short path; A=>F=>E=>D=>C is not shortest)
# "connected" means there is a path between any two nodes (no, because no A to X path)
# "cycle": path from node back to itself without repeats (A,B,C,D,E,F,A is a cycle, X,Y,X is not)
````

## Directed Graphs, DAGs

````python
from graphviz import Digraph # "di-rected graph"
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "A")
g
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "A")
g
# cyclic and connected (from any start, you can get anywhere)
# B is the child of A, C is the grandchild of A, A is the great grandchild of itself
# this is not a DAG because it is not acyclic (there's a cycle)
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("A", "C")
g.edge("C", "D")
g.edge("B", "D")
g
# no cycles (it is acyclic) and it is weakly connected, but not connected
# weakly connected: there's a path between any two nodes that IGNORES edge direction
# strongly connected (connected): there's a path between any two nodes that respects edge direction
# this is a DAG
````

````python
# A is a parent of B and C
# B is a child of A
# D is a grandchild of A
# B, C, and D are descendants of A
````

## Trees

**Tree**: DAG where one node (the **root**) has no parents and all others have exactly one parent

````python
g = Digraph()
g.edge("A", "B")
g.edge("A", "C")
#g.edge("C", "D")
g.edge("B", "D")
g.edge("D", "E")
g.edge("D", "F")
g.edge("D", "G")
g
# root: means node with zero parents
# leaf: means node with zero children
````

# Create Class(es) for a Binary Tree

````python
# class Edge:
#     pass # TODO: like edge weight, other details about the edge

class Node:
    def __init__(self, name):
        self.name = name
        
        # tree in general
        # self.children = [...list of other Node objects...]
        
        # linked list
        #self.child = ????
        
        # binary tree
        self.left = None
        self.right = None
        
    # prints out all the node names
    def dump(self, indent):
        print("    "*indent + self.name)
        
        if self.left != None:
            self.left.dump(indent+1)
            
        if self.right != None:
            self.right.dump(indent+1)
        
A = Node("A")
B = Node("B")
C = Node("C")
D = Node("D")
A.left = B
A.right = C
B.left = D # A.left.left = D

A.dump(0)
````
