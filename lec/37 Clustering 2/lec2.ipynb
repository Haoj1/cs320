```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.datasets import make_moons
from scipy.cluster.hierarchy import dendrogram
from IPython.core.display import Image
```

```python
df = gpd.read_file("counties.geojson")
df.head()
```

```python
df.plot()
```

```python
df.plot.scatter(x="forest", y="pasture", s=df["developed"]*200,
                c="crops", cmap="cool",
                sharex=False)
```

# Clustering Counties (KMeans)

```python
c = KMeans(4)
c.fit(df[["developed", "crops", "pasture", "forest"]])
groups = c.predict(df[["developed", "crops", "pasture", "forest"]])
print(c.inertia_)
df.plot(column=groups)
```

# Clustering Counties (Agglomerative)

```python
c = AgglomerativeClustering(4)

#c.fit(df[["developed", "crops", "pasture", "forest"]])
#groups = c.predict(df[["developed", "crops", "pasture", "forest"]])
groups = c.fit_predict(df[["developed", "crops", "pasture", "forest"]])

#print(c.inertia_)   # can't have inertia if you don't have centroids
df.plot(column=groups)
```

```python
df = pd.DataFrame(make_moons(noise=0.05)[0], columns=["x", "y"])
df["color"] = KMeans(n_clusters=2).fit_predict(df[["x", "y"]])
df.plot.scatter(x="x", y="y", c=df["color"], vmin=-1)
```
