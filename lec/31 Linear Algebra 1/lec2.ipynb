## Vector Multiplication: Overview

### Elementwise Multiplication

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

### Dot Product

$\begin{bmatrix}
1 & 2 & 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

```python
v1 = np.array([1,2,3])
v1
```

```python
v2 = np.array([4,5,6])
v2
```

```python
v1 * v2    # [1*4, 2*5, 3*6]
```

```python
v1 = np.array([1,2,3]).reshape(-1, 1) # vertical vector
v1
```

```python
v2 = np.array([4,5,6]).reshape(-1, 1) # vertical vector
v2
```

```python
v1*v2
```

## Transpose

```python
v1.T
```

```python
v1.T.T.T.T
```

## Broadcast

```python
print(v1.shape) # broadcasting stretches (3,1) => (3,3)
v1
```

```python
print(v2.T.shape) # broadcasting stretches (1,3) => (3,3)
v2.T
```

```python
v1*v2.T
```

```python
v1_broadcast = np.concatenate([v1, v1, v1], axis=1)
v1_broadcast
```

```python
v2_T_broadcast = np.concatenate([v2.T,v2.T,v2.T], axis=0)
v2_T_broadcast
```

```python
v1_broadcast * v2_T_broadcast
```
